## Erreur d&#8217;arrondi

*Exercice: Erreur d&#8217;arrondi*\
On considère les systèmes linéaires


$$
A_n \mathbf{x}_n=\mathbf{b}_n
$$

où $A_n$ est la matrice de Hilbert d&#8217;ordre $n$, définie par


$$
\left(A_n\right)_{i j}=\frac{1}{i+j-1}, \quad i, j=1, \ldots, n,
$$

et $\mathbf{b}_n$ est un vecteur tel que la solution du système (1) soit $\mathbf{x}_n=(1,1, \ldots, 1)^T$.

On veut étudier la propagation des erreurs d&#8217;arrondi sur la solution calculée par la méthode de factorisation LU avec pivoting. En général, on a le résultat suivant: pour la solution $\mathbf{x}_n+\boldsymbol{\delta} \mathbf{x}$ du système perturbé


$$
A_n\left(\mathbf{x}_n+\boldsymbol{\delta} \mathbf{x}\right)=\left(\mathbf{b}_n+\boldsymbol{\delta} \mathbf{b}\right)
$$

on a


$$
E_n=\frac{\|\boldsymbol{\delta} \mathbf{x}\|}{\left\|\mathbf{x}_n\right\|} \leq K_n \frac{\|\boldsymbol{\delta} \mathbf{b}\|}{\left\|\mathbf{b}_n\right\|},
$$

où $K_n$ est le conditionnement de $A_n$, et $\|\cdot\|$ est la [norme euclidienne](python:vector.ipynb#_vector_norms).
