# Série 10: Problèmes aux valeurs propres

## Exercices

*Exercice 1*\
En prenant une tolérance de $\varepsilon=10^{-10}$ et en partant de la donnée initiale $\mathbf{x}^{(0)}=(1,2,3)^T$, utiliser la méthode de la puissance pour approcher la valeur propre de module maximal des matrices suivantes


$$
A_1=\left[\begin{array}{lll}
1 & 2 & 0 \\
1 & 0 & 0 \\
0 & 1 & 0
\end{array}\right], A_2=\left[\begin{array}{ccc}
0.1 & 3.8 & 0 \\
1 & 0 & 0 \\
0 & 1 & 0
\end{array}\right], A_3=\left[\begin{array}{ccc}
0 & -1 & 0 \\
1 & 0 & 0 \\
0 & 1 & 0
\end{array}\right] .
$$

Commenter la convergence de la méthode dans les trois cas.
