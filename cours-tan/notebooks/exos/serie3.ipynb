$$
a_{ij}=0 \quad \textrm{pour} \quad |i-j|\geq p\Rightarrow\left\{
\begin{matrix}
l_{ij}=0 \quad \textrm{pour} \quad i-j\geq p,\\[2mm]
u_{ij}=0 \quad \textrm{pour} \quad j-i\geq p.
\end{matrix}\right.
$$
