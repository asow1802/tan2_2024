## Déformations d&#8217;un tissu biologique


![disque](interpolation/disque-intervertebral.png)

*Déformations d&#8217;un tissu biologique*\
On considère un test mécanique pour établir le lien entre contrainte $\sigma$ ($MPa=100 N/cm^2$) et déformations relatives $\epsilon$ (cm/cm) d&#8217;un échantillon de tissu biologique (disque intervertébral), voir la [figure](#fig:disque).

Ref: P. Komarek, Ch. 2 de *Biomechanics of Clinical Aspects of Biomedicine*, 1993, J. Valenta ed., Elsevier

* **Mise en traction**\
On effectue un ensemble de tests: on impose une Force $F$, on calcule la contrainte $\sigma$ et la déformation $\epsilon$

  
  | Test | contrainte | déformation |
  | ---- | ---------- | ----------- |
  | $i$ | $\sigma$ | $\epsilon$ |
  | 1 | 0.00 | 0.00 |
  | 2 | 0.06 | 0.08 |
  | 3 | 0.14 | 0.14 |
  | 4 | 0.25 | 0.20 |
  | 5 | 0.31 | 0.23 |
  | 6 | 0.47 | 0.25 |
  | 7 | 0.60 | 0.28 |
  | 8 | 0.70 | 0.29 |
  



A partir des données de la [table](#tab:3.2), on veut estimer la déformation correspondante à un effort $\sigma=0.9$ MPa. ?
Une solution sera d&#8217;utiliser la méthodes des moindres carrés

![image](interpolation/regression-disque.jpg)

A travers la méthode des moindres carrés on obtient que la droite qui approche mieux ces données est $p(x)=0.3938 x -0.0629$. On peut utiliser cette droite (dite *de régression*) pour estimer $\epsilon$ lorsque $\sigma=0.9$ MPa: on trouve $p(0.9) \simeq 0.4$, voir la [figure](#fig:3.3).

## Débit sanguin dans la carotide


![image](interpolation/carotid.jpg)

Les points dans la figure représentent les mesures du débit du sang dans une section de l&#8217;artère carotide pendant un battement cardiaque. La fréquence d&#8217;acquisition de données est constante et égale à $10/T$ où $T=1 s$. est la période du battement.
On veut déduire de ce signal discret un signal continu représenté par une combinaison linéaire de fonctions connues (par exemple de fonctions trigonométriques, cette représentation est adaptée a notre problème parce que la fonction obtenue va bien approximer un signal périodique).
Débit sanguin dans la carotide: signal Soit $f(t)$ le signal dont on connaı̂t $N = 10$ échantillons $[f(t_0),\ldots,f(t_{N-1})$], avec $t_j = j T/N$. On cherche $\{c_k\} \in {\mathbb C}$, $k \in [0,N-1$\subset{\mathbb N}] tels que:

$$
    f(t_j)=\frac{1}{N}\sum_{k=0}^{N-1} c_k
    % \text{avec} \quad \omega_N=e^{(-2\pi i)/N}.
    \omega_N^{- k j }, \quad j = 0, \ldots, N-1,
$$
où

$$
\omega_N=e^{(-2\pi i)/N} = \cos(2\pi/N) - i \sin(2\pi/N),
$$
étant $i$ l&#8217;unité imaginaire.
*Note:* On peut calculer le vecteur des coefficients ${\bf c}=[c_1,\ldots,c_{10}$^T] par l&#8217;algorithme F.F.T. (Fast Fourier Transform, Cooley et Tuckey, 1965), implémenté dans Numpy avec la commande `fft`.
A partir de l'[expression](#eq:interp:fft), il y a plusieurs techniques pour reconstruire la valeur de $f$ dans chaque point $t\in[0,T$]. Une fois que l&#8217;on a reconstruit la fonction, on peut la représenter sur un graphe comme celui de la [figure](#fig:3.6).
