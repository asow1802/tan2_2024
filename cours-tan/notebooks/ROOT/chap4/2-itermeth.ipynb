## Méthodes itératives pour les systèmes linéaires

L&#8217;idée des méthodes itératives est de construire une suite de vecteurs $\mathbf{x}^{(k)}$ qui *converge* vers le vecteur $\mathbf{x}$, solution du système $A\mathbf{x}=\mathbf{b}$,

$$
\mathbf{x}=\lim_{k\to\infty}\mathbf{x}^{(k)}.
$$
Une technique générale pour déduire des méthodes itératives est basée sur une décomposition (*splitting*) de la matrice $A$ sous la forme $A=P-N$, où $P$ et $N$ sont des matrices à déterminer où $P$ est non singulière. La matrice $P$ est appelée *la matrice de préconditionnement*.
Plus précisément, si $\mathbf{x}^{(0)}$ est donné, on peut calculer $\mathbf{x}^{(k+1)}$ pour $k\geq 0$ solution du système

$$
\label{itt}
P\mathbf{x}^{(k+1)}=N\mathbf{x}^{(k)}+\mathbf{b},\,\,\,\,k\geq 0.
$$
Clairement, la solution exacte $\mathbf{x}$ satisfait latexmath:[P\mathbf{x} =
N\mathbf{x} + \mathbf{b}] (*consistance* de la méthode).
Le système [ci-dessus](#eq2) peut être écrit également sous la forme

$$
\label{harry}
\mathbf{x}^{(k+1)}=B\mathbf{x}^{(k)}+P^{-1}\mathbf{b},\,\,\,\,k\geq 0,
$$
où $B=P^{-1}N$ est la *matrice d&#8217;itération* de la méthode. En définissant l&#8217;erreur au pas $k$ comme

$$
\mathbf{e}^{(k)} = \mathbf{x}^{(k)}-\mathbf{x},
$$
on obtient la relation de récurrence (équation de l&#8217;erreur)

$$
\mathbf{e}^{(k)}=B\mathbf{e}^{(k-1)}, \quad \text{et donc }
\mathbf{e}^{(k)}=B^k\mathbf{e}^{(0)}, \quad k=0,1,\ldots.
$$
On a le résultat suivant.
*Théorème: Convergence des méthodes itératives*\
Si la méthode [itérative](#eq3) est consistante, la suite de vecteurs $\{ \mathbf{x}^{(k)} \}$ de [[eq3]](#eq3) converge vers la solution du système A$\mathbf{x} = \mathbf{b}$ pour toute donnée initiale $\mathbf{x}^{(0)}$ si et seulement si $\rho(B)<1$.
* **Soit $\rho(B)<1$**\
On sait que $\lim_{k \to \infty} \textrm{A}^k = 0 \Leftrightarrow \rho(\textrm{A})<1$ (voir p27 Livre Quarteroni Sacco).
Donc, $\rho(B)<1 \Rightarrow \lim_{k \to \infty} B^k = 0$ $\Rightarrow$ $\lim_{k \to \infty} B^k \mathbf{e}^{(0)} = 0$ pour tout $\mathbf{e}^{(0)} \in \mathbb{R}^n$.
Mais $B^k \mathbf{e}^{(0)} = \mathbf{e}^{(k)}$ $\Rightarrow$ $\lim_{k \to \infty} \mathbf{e}^{(k)} = 0$.
* **Soit maintenant $\rho(B) \geq 1$**\
alors il existe une valeur propre $\lambda^*$ telle que $|\lambda^*|\geq 1$. Soit $\mathbf{e}^{(0)}$ le vecteur propre associé à $\lambda^*$. Alors, $B \mathbf{e}^{(0)} = \lambda^* \mathbf{e}^{(0)}$ et donc $\mathbf{e}^{(k)} = (\lambda^*)^k \mathbf{e}^{(0)}$ ce qui implique $\lim_{k \to \infty} \mathbf{e}^{(k)} \not= 0$. $\Box$
