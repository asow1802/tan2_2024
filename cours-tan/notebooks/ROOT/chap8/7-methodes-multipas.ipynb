# Les Méthodes multi-pas

Les méthodes multi-pas sont des méthodes qui utilisent plusieurs valeurs précédentes pour calculer la valeur suivante contrairement aux [méthodes à un pas](chap8/7-methodes-unpas.ipynb) qui n&#8217;utilisent que la valeur précédente.


Elles sont souvent utilisées pour résoudre des équations différentielles ordinaires.
Les méthodes multi-pas sont souvent plus précises que les méthodes à un pas, mais elles sont aussi plus compliquées à implémenter.


Les méthodes multi-pas atteignent un ordre élevé en utilisant plusieurs valeurs $u_n, u_{n-1}, \ldots$, pour construire $u_{n+1}$.
Ces méthodes sont généralement obtenues à partir de la formule suivante (il suffit d&#8217;intégrer [le problème](chap8/1-intro.ipynb#prob:cauchy)  sur $\left[t_n, t_{n+1}\right]$ )



$$
y_{n+1}=y_n+\int_{t_n}^{t_{n+1}} f(t, y(t)) d t,
$$


en approchant l&#8217;intégrale par une formule de quadrature interpolatrice appropriée.


* **Adams-Bashforth d&#8217;ordre 2**\
Si on remplace $f$ dans [l&#8217;équation](#eq:451) par son polynôme d&#8217;interpolation de degré $1$ $\Pi_1 f$ aux noeuds $t_{n-1}, t_n$ on obtient la méthode (à 2 pas explicite) d&#8217;Adams-Bashforth d&#8217;ordre 2 (AB2) :

  
$$
u_{n+1}=u_n+\frac{h}{2}\left[3 f_n-f_{n-1}\right] .
$$


* **Adams-Bashforth d&#8217;ordre 3**\
Si on remplace $f$ dans [l&#8217;équation](#eq:451) par son polynôme d&#8217;interpolation de degré $2$ $\Pi_2 f$ aux noeuds $t_{n-2}, t_{n-1}, t_n$ on obtient la méthode (à 3 pas explicite) d&#8217;Adams-Bashforth d&#8217;ordre 3 (AB3):

  
$$
u_{n+1}=u_n+\frac{h}{12}\left[23 f_n-16 f_{n-1}+5 f_{n-2}\right] .
$$

  Cette méthode est illustré ci-après.


* **Adams-Moulton d&#8217;ordre 3**\
En remplaçant $f$ dans [l&#8217;équation](#eq:451) par son polynôme d&#8217;interpolation de degré $2$ $\Pi_2 f$ aux noeuds $t_{n-1}, t_n, t_{n+1}$, on obtient la méthode (à 2 pas implicite) d&#8217;Adams-Moulton d&#8217;ordre 3 (AM3)

  
$$
u_{n+1}=u_n+\frac{h}{12}\left[5 f_{n+1}+8 f_n-f_{n-1}\right] .
$$

  Cette méthode est illustré ci-après


* **Adams-Moulton d&#8217;ordre 4**\
En remplaçant $f$ dans [l&#8217;équation](#eq:451) par son polynôme d&#8217;interpolation de degré $3$ $\Pi_3 f$ aux noeuds $t_{n-2}, t_{n-1}, t_n, t_{n+1}$, on obtient la méthode (à 3 pas implicite) d&#8217;Adams-Moulton d&#8217;ordre 4 (AM4)



$$
u_{n+1}=u_n+\frac{h}{24}\left[9 f_{n+1}+19 f_n-5 f_{n-1}+f_{n-2}\right] .
$$

## Backward Differences

Une autre famille de méthodes multi-pas peut être conçue en écrivant [l&#8217;équation](chap8/1-intro.ipynb#eq:cauchy)  en $t_{n+1}$ et en remplaçant $y^{\prime}\left(t_{n+1}\right)$ par un quotient incremental d&#8217;ordre élevé. En particulier, on peut dériver ainsi la méthode BDF (Backward Difference Formula) implicite d&#8217;ordre 3 suivante :

$$
u_{n+1}=\frac{18}{11} u_n-\frac{9}{11} u_{n-1}+\frac{2}{11} u_{n-2}+\frac{6 h}{11} f_{n+1}
$$
