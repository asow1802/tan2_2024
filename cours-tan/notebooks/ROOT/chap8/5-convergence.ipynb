# Convergence

## Convergence d&#8217;Euler progressif

*Définition: Convergence et ordre de convergence*\
Soit $y(t)$ la solution du [problème de Cauchy](chap8/1-intro.ipynb#cauchy)  sur l&#8217;intervalle $[0,T]$;
soit $u_n$ une solution approchée au temps $t_n = nh$, où $h = T/N_h$ (latexmath:[N_h \in \mathbb{N}\]) est le pas de temps, trouvée par une méthode numérique donnée.

La méthode est dite *convergente* si


$$
\begin{aligned}
    \forall n=1,\ldots,N_h: \quad | u_n - y(t_n)| \leq C(h)

\end{aligned}
$$

où $C(h)\to 0$ lorsque $h\to 0$.

Si en plus il existe $p>0$ tel que $C(h) = \mathcal{O}(h^p)$, on dit que la méthode est *convergente d&#8217;ordre* $p$.
Dans la suite on va analyser la convergence et l&#8217;ordre de la méthode d&#8217;Euler progressive.
L&#8217;analyse de la convergence de la méthode d&#8217;Euler retrograde est similaire.
On va démontrer le résultat de convergence suivant :
*Théorème: convergence de la méthode d&#8217;Euler progressive*\
Si $y\in \mathcal{C}^2([0,T])$ et $f$ est lipschitzienne, $L$ étant sa constante de Lipschitz, alors


$$
\label{thm}
\forall n\geq 0,\,\,\,|y(t_n)-u_n| \leq c(t_n)h,
$$

où


$$
c(t_n)=\frac{e^{Lt_n}-1}{2L}\max_{t\in[0,T]} |y''(t)|.
$$

En particuler, la méthode est convergente d&#8217;ordre $p=1$ selon la définition précédente, avec


$$
C(h) = c(T)h.
$$
On définit l&#8217;erreur au pas $n$ par :


$$
e_n=u_n - y(t_n)
$$


![image](ode/ode_en.svg)


On définit l'*erreur de troncature locale* de la méthode d&#8217;Euler progressive comme


$$
    \tau_{n+1}(h) = \frac{y(t_{n+1})-y(t_n)}{h}- y'(t_n),
$$

et l'*erreur de troncature globale*


$$
\tau(h) = \max_n |\tau_n(h)|.
$$

On sait qu&#8217;il existe $\eta_n \in (t_n,t_n+h)$ tel que


$$
\tau_{n+1}(h) = \frac{1}{2}y''(\eta_n)h.
$$

Donc on a l&#8217;estimation suivante pour erreur de troncature globale:


$$
\tau(h) \leq \frac{1}{2}\max_{t\in[0,T]}|y''(t)| h.
$$

De l&#8217;équation suivante pour la solution numérique $u_n$


$$
\left\{
\begin{array}{l}
\displaystyle\frac{u_{n+1}-u_n}{h}=f(t_n,u_n)\qquad \text{pour }\,n=0,1,2\ldots\\[2mm]
u_0=y_0.
\end{array}
\right.
$$

et [l&#8217;équation](#erreurtronc2) pour l&#8217;erreur de troncature locale


$$
\tau_{n+1}(h) = \frac{y(t_{n+1})-y(t_n)}{h}-f(t_n,y(t_n)) = \frac{y(t_{n+1})-y(t_n)}{h} - y'(t_n),
$$

on obtient:


$$
\left\{
\begin{array}{l}
\displaystyle\frac{e_{n+1}-e_n}{h}= f(t_n,u_n) - f(t_n,y(t_n)) - \tau_{n+1}(h),\\[2mm]
e_0=0.
\end{array}
\right.
$$

Comme la fonction $f$ est lipschitzienne, on a


$$
|f(t_n,u_n) - f(t_n,y(t_n))| \leq  L |u_n - y(t_n)| \leq L |e_n|.
$$

Compte tenu de cette inégalité, [l&#8217;équation sur l&#8217;erreur](#erreur) permet d&#8217;écrire que :


$$
|e_{n+1}| \leq (1+Lh) |e_n| + h |\tau_{n+1}(h)|.
$$

Soit $E_j= |e_j|$. Alors on a les inégalités suivantes :


$$
\begin{aligned}
E_{n+1} &\leq &(1+hL)E_n+h\tau(h)\\
&\leq& (1+hL)\left[(1+hL)E_{n-1}+h\tau(h)\right]+h\tau(h)\\
&\leq& \left[1+(1+hL)+(1+hL)^2+\ldots+(1+hL)^n\right]h\tau(h)\\
&=& \frac{(1+hL)^{n+1}-1}{hL}h\tau(h)
\end{aligned}
$$

Mais


$$
1+hL\leq e^{hL}, \;\forall h>0,
$$

d&#8217;où


$$
(1+hL)^n-1\leq e^{Lhn}-1=e^{Lt_n}-1.
$$

Donc


$$
E_n \leq   \frac{e^{Lt_n}-1}{L}\tau(h)\leq\frac{e^{Lt_n}-1}{L}h\frac{1}{2}\max_{t\in [0,T]} |y''(t)|,
$$

c&#8217;est-à-dire


$$
|u_n - y(t_n)| \leq \left[
\frac{1}{2} \frac{e^{Lt_n} - 1}{L} \max_{t\in [0,T]} |y''(t)|
\right] h
\leq c(T) h,
$$

où $c(T) = \frac{e^{LT} - 1}{L} \max_{t\in [0,T]} |y''(t)|$.
*Note:* *Le même type de résultat peut être établi pour la méthode d&#8217;Euler rétrograde.*
*Note:* [L&#8217;estimation de convergence](#thm:convergence-euler) est obtenue sous l&#8217;hypothèse que la fonction $f$ soit continue et lipschitzienne. Une estimation plus précise, c-à-d

$$
  |u_n - y(t_n)| \leq h t_n \frac{1}{2} \max_t |y''(t)| \, ,
$$
est vraie si $f$ satisfait aussi la condition
latexmath:[\frac{\partial
  f}{\partial y} (t,y) \leq 0] pour tout $t \in [0,T$] et pour
tout latexmath:[y
\in (-\infty, \infty)]._
*Montrons [l&#8217;équation plus précise](#plusprecise).*\
D&#8217;après le théorème de Lagrange, il existe $\xi_n$ tel que


$$
f(t_n,u_n) - f(t_n,y(t_n)) = \frac{\partial f (t,\xi_n)}{\partial y} (u_n - y(t_n)) =
\frac{\partial f (t,\xi)}{\partial y} e_n.
$$

Donc, à partir de [[erreur]](#erreur) on trouve


$$
e_{n+1} = \left(1 + h \frac{\partial f (t,\xi_n)}{\partial y}\right) e_n
- h \frac h2 y''(\eta_n).
$$

Si $h<\frac{1}{\lambda_{max}}$ on a
$1 + h \frac{\partial f (t,\xi_n)}{\partial y} \in (-1,1)$, et
donc


$$
| e_{n+1} | \leq |e_n| + h\tau(h).
$$

Comme $e_0 = 0$, on tire


$$
|e_n| \leq nh\tau(h) = t_n \tau(h),
$$

c.-à-d. on trouve [[plusprecise]](#plusprecise).
*Important:* On peut donc conclure que la méthode d&#8217;Euler progressive et retrograde sont convergentes d&#8217;ordre 1.
Pour illustrer la convergence des 2 méthodes d&#8217;Euler, on va résoudre le problème de Cauchy suivant:
*Exemple: Convergence des méthodes d&#8217;Euler progressive et retrograde*\
Résoudre le problème de Cauchy suivant:


$$
\begin{aligned}

\left\{ \begin{array}{ll}
y^\prime = 1 - y^2, & t > 1, \\[3mm]
y(1) = (e^{2}-1)/(e^{2}+1). &
\end{array} \right.
\end{aligned}
$$

La solution exacte de ce problème est $y(t) = (e^{2t}-1)/(e^{2t}+1)$.


```python
import numpy as np
from tan.ode.feuler import feuler
from tan.ode.beuler import beuler

def f(t,y):
    return 1 - y**2

def yexact(t):
    return (np.exp(2*t)-1)/(np.exp(2*t)+1)

Tspan = [0,2]
results = []
orders = []
for N in [10,20,40,80,160]:
    h = (Tspan[1]-Tspan[0])/N
    y0 = [0]
    tf,yf = feuler(f, Tspan, y0,N)
    tb,yb = beuler(f, Tspan, y0,N)
    y = yexact(tf)
    results.append([h,yf[-1,0],yb[-1,0],y[-1],abs(yf[-1,0]-y[-1]),abs(yb[-1,0]-y[-1])])
    if len(results) > 1:
        orders.append([h,np.log(results[-2][4]/results[-1][4])/np.log(2),np.log(results[-2][5]/results[-1][5])/np.log(2)])

from tabulate import tabulate
print(tabulate(results, headers=['h', 'Euler progressif', 'Euler retrograde', 'Exact', 'Erreur Euler progressif', 'Erreur Euler retrograde'], tablefmt='grid'))
print(tabulate(orders, headers=['h', 'Ordre de convergence (progressive)', 'Ordre de convergence (rétrograde)'], tablefmt='grid'))
```
