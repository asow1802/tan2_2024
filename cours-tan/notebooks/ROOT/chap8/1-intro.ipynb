# Introduction

## Objectifs

On considère une fonction continue $f:\mathbb{R}_+\times\mathbb{R}\rightarrow\mathbb{R}$.
*Problème de Cauchy*\
Pour $y_0\in\mathbb{R}$ donné, on cherche $y:t\in\mathbb{R}_+\rightarrow y(t)\in\mathbb{R}$ qui satisfait le problème suivant, appelé *problème de Cauchy*:


$$
\left\{
\begin{array}{l}
y'(t)=f(t,y(t))\qquad \text{si }\,t>0\\
y(0)=y_0
\end{array}
\right.
$$

où $y'(t)=\displaystyle\frac{d y(t)}{d t}$.

## Exemples

* **Un problème de Cauchy  linéaire**\
comme par exemple:

  
$$
\left\{
\begin{array}{l}
y'(t)=3y(t)-3t\qquad \text{si }\,t>0\\
y(0)=1
\end{array}
\right.
$$

  pour lequel $f(t,v)=3v-3t$ et dont la solution est
  $y(t)=(1-1/3)e^{3t}+t+1/3$.


* **Problèmes non-linéaires**\
comme:

  
$$
\left\{
\begin{array}{l}
y'(t)=\sqrt[3]{y(t)}\qquad \text{si }\,t>0\\
y(0)=0
\end{array}
\right.
$$

  avec $f(t,v)=\sqrt[3]{v}$. Ce problème admet les trois solutions suivantes : $y(t)=0$, $y(t)=\sqrt{8t^3/27}$, $y(t)=-\sqrt{8t^3/27}$.


* **Problème avec solution locale**\
par exemple

  
$$
\left\{
\begin{array}{l}
y'(t)=1+y^2(t)\qquad \text{si }\,t>0\\
y(0)=0
\end{array}
\right.
$$

  admet comme solution la fonction $y(t)=\tan(t)$
  $0<t<\frac{\pi}{2}$, c&#8217;est-à-dire une solution locale.


*Théorème de Cauchy-Lipschitz*\
Si


- $f$ est continue sur $\mathbb{R}_+\times\mathbb{R}$ et
- il existe une constante $L>0$ telle que

$$
\label{lipschitz}
    |f(t,v)-f(t,w)| \leq L |v-w| \quad
           \forall v,w\in\mathbb{R}, \; \forall t>0,
$$


alors

le [problème de Cauchy](#prob:cauchy) admet une solution globale (c-à-d. pour tout $t>0$) et elle est unique.
